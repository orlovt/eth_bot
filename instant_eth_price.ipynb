```python
from pycoingecko import CoinGeckoAPI
```

```python
class prices():
    def get_eth_usd():
        cg = CoinGeckoAPI()
        return 'ETH/USD: \t' + str(cg.get_price(ids='ethereum', vs_currencies='usd').get('ethereum').get('usd'))
    def get_eth_rub():
        cg = CoinGeckoAPI()
        return 'ETH/RUB: \t' + str(cg.get_price(ids='ethereum', vs_currencies='rub').get('ethereum').get('rub'))
    def get_usd_rub():
        cg = CoinGeckoAPI()
        return 'USD/RUB: \t' + str(cg.get_price(ids='busd', vs_currencies='rub').get('busd').get('rub'))
```

```python
print(prices.get_eth_usd())
print(prices.get_eth_rub())
print(prices.get_usd_rub())
```
